```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
'''
Prep course
U3.L3

Learning Drills
'''

# P1

# n = 1000, mean = 0, std = 1
mean = 0
std = 1
n = 1000

df = pd.DataFrame({'rand': np.random.normal(mean, std, n)})

df.hist()
plt.show()
```

```python
# P2

# quantile-quantile distribution to visualize normality of random variable distributions
rand1 = np.random.normal(50, 300, 1000)
rand2 = np.random.poisson(1, 1000)

rand1.sort()
rand2.sort()

# normally distribution control variable
norm = np.random.normal(0, 1, 1000)

norm.sort()

# rand1 against norm
plt.plot(norm, rand1, 'o')
plt.show()

# rand2 against norm
plt.plot(norm, rand2, 'o')
plt.show()
```

```python
# histograms to compare

def plot_histogram(data):
    plt.hist(data, bins=20, color='c')

    # vertical line at mean
    plt.axvline(data.mean(), color='b', linestyle='solid', linewidth=2)

    # vertical line at 1 std above mean
    plt.axvline(data.mean() + data.std(), color='b', linestyle='dashed', linewidth=2)

    # vertical line at 1 std below mean
    plt.axvline(data.mean() - data.std(), color='b', linestyle='dashed', linewidth=2)

    plt.show()

plot_histogram(rand1)
plot_histogram(rand2)
```

```python
# P3

# Bernoulli distribution
bernoulli = np.random.binomial(1, .5, 1000)

plt.hist(bernoulli)
plt.show()

# Binomial distribution n=20, p=.5
binomial = np.random.binomial(20, .5, 1000)

plt.hist(binomial)
plt.show()

# Gamma distribution shape=5, scale=1
gamma = np.random.gamma(5, 1, 1000)

plt.hist(gamma)
plt.show()

# Poisson distribution lambda=3
poisson = np.random.poisson(3, 1000)

plt.hist(poisson)
plt.show()
```

```python
# Conditional Data
ecommerce = pd.DataFrame()
ecommerce['spending'] = np.random.normal(75, 25, 1000)

plt.hist(ecommerce['spending'])
plt.show()

ecommerce['visit_count'] = np.random.randint(0, 5, 1000)

plt.hist(ecommerce[ecommerce['visit_count'] > 2]['spending'])
plt.show()
```
