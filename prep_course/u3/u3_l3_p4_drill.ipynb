```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
'''
Prep Course
U3.L3.P4

task: - create 6 datasets and statistically summarize each with a unique distribution
      - visualizations should include mean and std plotted with vertical lines
      - identify whether the descriptive statistics provided useful information - are there common characteristics
        to those which could be usefully described?
      - generate 2 normally distributed variables with predefined params, add them to form and graph a third the compute
        and evaluate the descriptive stats against the data
'''

# Data set 1 - Binomial distribution - 100 events at p=.25 over 1000 trials
data1 = np.random.binomial(100, .25, 1000)

plt.axvline(data1.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(data1.mean() + data1.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(data1.mean() - data1.std(), color='r', linestyle='dashed', linewidth=2)

plt.hist(data1)
plt.show()
```

```python
# Data set 2 - Gamma distribution - shape=2, scale=2 over 1000 trials => mean=2*2 and std=2
data2 = np.random.gamma(2, 2, 1000)

plt.axvline(data2.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(data2.mean() + data2.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(data2.mean() - data2.std(), color='r', linestyle='dashed', linewidth=2)

plt.hist(data2)
plt.show()
```

```python
# Data set 3 - Poisson distribution - lambda = frequency of occurrence = 10 over 1000 trials
data3 = np.random.poisson(10, 1000)

plt.axvline(data3.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(data3.mean() + data3.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(data3.mean() - data3.std(), color='r', linestyle='dashed', linewidth=2)

plt.hist(data3)
plt.show()
```

```python
# Data set 4 - Geometric distribution - p=.5 over 1000 trials
# Geometric distribution - Bernoulli trials are events with binary outcomes - Geometric distributions model
# the number of trials that must be run to achieve a success state - ie. if p=.5, mean should be ~2
data4 = np.random.geometric(.5, 1000)

plt.axvline(data4.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(data4.mean() + data4.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(data4.mean() - data4.std(), color='r', linestyle='dashed', linewidth=2)

plt.hist(data4)
plt.show()
```

```python
# Data set 5 - Logistic distribution - loc=10, scale=1 over 1000 trials
# Logistic distribution - uses a logistic function to model a categorical outcome and used for logistic regression
data5 = np.random.logistic(10, 1, 1000)

plt.axvline(data5.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(data5.mean() + data5.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(data5.mean() - data5.std(), color='r', linestyle='dashed', linewidth=2)

plt.hist(data5)
plt.show()
```

```python
# Data set 6 - Normal distribution - loc=10, scale=10 over 1000 trials
data6 = np.random.normal(10, 10, 1000)

plt.axvline(data6.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(data6.mean() + data6.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(data6.mean() - data6.std(), color='r', linestyle='dashed', linewidth=2)

plt.hist(data6)
plt.show()

# In most cases the descriptive statistics are useful indicators of the underlying data
# Distributions like the Poisson distribution less so - as the mean isn't the mode and there are two peaks
# surrounding - however the standard deviation does provide valid indication of the variability
```

```python
# Additional drill
var1 = np.random.normal(5, .5, 1000)
var2 = np.random.normal(10, 1, 1000)
var3 = var1+var2

plt.axvline(var3.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(var3.mean() + var3.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(var3.mean() - var3.std(), color='r', linestyle='dashed', linewidth=2)

plt.hist(var3)
plt.show()

# since we added the two datapoints together our mean is now the sum of the two individual means.  
# the standard deviation is slightly increased
```
